# **Classification**

```
# Import libraries
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load dataset
data = load_breast_cancer()
X = data.data  #features
y= data.target #target

# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state =667)

# Defines and builds the lazyclassifier
clf = LazyClassifier(ignore_warnings=True, custom_metric=None,predictions=True)
#models_train,predictions_train = clf.fit(X_train, X_train, y_train, y_train)
models_c, predictions_c = clf.fit(X_train, X_test, y_train, y_test)
```

```
predictions_c
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=models_c.index, y="Accuracy", data=models_c,hue=models_c.index)
plt.xticks(rotation=90)
```

# **Regression**

```
# Import libraries
from lazypredict.Supervised import LazyRegressor
from sklearn.utils import shuffle
import numpy as np
import pandas as pd

# Load dataset
path="https://frenzy86.s3.eu-west-2.amazonaws.com/IFAO/boston_houses.csv"
df = pd.read_csv(path)

X = df.drop(columns='MEDV')
y = df['MEDV']

# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state =667)

# Defines and builds the lazyclassifier
reg = LazyRegressor(ignore_warnings=False, custom_metric=None,predictions=True)
models_r,predictions_r = reg.fit(X_train, X_test, y_train, y_test)
```
