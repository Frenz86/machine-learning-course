# **Classification**

```
# Import libraries
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Load dataset
data = load_breast_cancer()
X = data.data
y= data.target

# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state =667)

# Defines and builds the lazyclassifier
clf = LazyClassifier(ignore_warnings=True, custom_metric=None,predictions=True)
#models_train,predictions_train = clf.fit(X_train, X_train, y_train, y_train)
models_c, predictions_c = clf.fit(X_train, X_test, y_train, y_test)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 10))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=models_c.index, y="Accuracy", data=models_c)
plt.xticks(rotation=90)
```

# **Regression**

```
# Import libraries
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np

# Load dataset
boston = datasets.load_boston()
X, y = shuffle(boston.data, boston.target, random_state=667)

# Data split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state =667)

# Defines and builds the lazyclassifier
reg = LazyRegressor(ignore_warnings=False, custom_metric=None,predictions=True)
models_r,predictions_r = reg.fit(X_train, X_test, y_train, y_test)
```
