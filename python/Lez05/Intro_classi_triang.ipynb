```python
class Triangolo:

    """
    questa classe rappresenta il triangolo
    """
    def __init__(self,a,b,c,h):
        self.a = a
        self.b = b
        self.c = c
        self.h = h

    def area(self):
        """
        Calcolo area
        """
        at = (self.b*self.h)/2
        return at

    def perimetro(self):
        """
        Calcolo perimetro
        """
        per = self.a+self.b+self.c
        return per

    def print_info(self):
        print("Area del triangolo = %2.f"% self.area())
        print("Perimentro del triangolo = %2.f"% self.perimetro())
```

```python
help(Triangolo)
```

```python
trian =  Triangolo(6,5,4,4)
```

```python
trian.area()
```

```python
trian.perimetro()
```

```python
trian.print_info()
```
